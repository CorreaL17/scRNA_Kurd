```python
#Use scanoy concatenate function to merge everythign into one object
Combined=sc.concat([Naive,D3_SPL,D4_SPL,D5_SPL,D6_SPL,D7_SPL,D7_SPL2,D10_SPL,D10_SPL2,D14_SPL,D14_SPL2,D21_SPL,D21_SPL2,D32_SPL,D60_SPL,D90_SPL],join="outer",label="dataset")
```

```python
#Filter low count genes/cells
sc.pp.filter_cells(Combined, min_genes=200)
sc.pp.filter_genes(Combined, min_cells=3)
```

```python
#Removing mitochondrial reads. Since this data set had already been processes there are none
#Note mitochondrial reads in this data set are labeled with mt.
Combined.var['mt'] = Combined.var_names.str.startswith('mt.')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(Combined, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```

```python
sc.pl.violin(Combined, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'],
             jitter=0.4, multi_panel=True)
```

```python

sc.pl.scatter(Combined, x='total_counts', y='pct_counts_mt')
sc.pl.scatter(Combined, x='total_counts', y='n_genes_by_counts')
```

```python
Combined = Combined[Combined.obs.n_genes_by_counts < 2500, :]
Combined = Combined[Combined.obs.pct_counts_mt < 5, :]
```

```python
sc.pp.normalize_total(Combined)
sc.pp.normalize_per_cell(Combined)
```

```python

sc.pp.log1p(Combined)
```

```python
sc.pp.highly_variable_genes(Combined, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
sc.pl.highly_variable_genes(Combined)
```

```python
Combined.raw = Combined
```

```python
Combined = Combined[:, Combined.var.highly_variable]
```

```python
sc.pp.regress_out(Combined, ['total_counts', 'pct_counts_mt'])
```
